```python
import pandas as pd
import seaborn as sns

# Create a sample DataFrame
data = {
    'top_products': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E'],
    'Total_in_wishlist': [5, 8, 12, 4, 3],
    'Average_rating': [4.5, 3.8, 4.2, 4.9, 3.1],
    'Total_recommended_count': [12, 10, 15, 18, 6]
}
df = pd.DataFrame(data)

# Set the product names as the index
df = df.set_index('top_products')

# Create a colorful pairgrid
sns.pairplot(df, diag_kind='kde')

# Add labels to the diagonal plots
for i, ax in enumerate(sns.pairplot(df).axes.flat):
    if i % 4 == 0:
        ax.set_ylabel('Total Items in Wishlist')
    elif i % 4 == 1:
        ax.set_ylabel('Average Rating')
    elif i % 4 == 2:
        ax.set_ylabel('Total Recommended Count')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming your data is stored in a Pandas DataFrame called `df`:

# Filter data for Sheet Mask category
sheet_mask_data = ds1[ds1['Default_category'] == 'Sheet Mask']

# Group by 'Product_name' and calculate sum of 'Total_repurchase_yes_count'
product_counts = sheet_mask_data.groupby('Product_name')['Total_repurchase_yes_count'].sum()

# Sort by descending order and select top 5 products
top_5_products = product_counts.sort_values(ascending=False).head(5)

# Prepare data for bar chart
product_names = top_5_products.index.to_list()
repurchase_counts = top_5_products.values.tolist()

# Create the bar chart
plt.bar(product_names, repurchase_counts)

# Add labels and title
plt.xlabel('Product Names')
plt.ylabel('Total Repurchases')
plt.title('Most Repurchased Sheet Masks')
plt.xticks(rotation=45, ha='right')
# Display the chart
plt.show()
```
